```python
import pandas as pd
import numpy as np
from datetime import datetime
```

MetaData
--

1. items : items redeemed in each redemption order
2. k_user_id : all the merchants ids
3. user_id : all the users id
4. paid_amount: transaction value for the redemption
5. paid_at: epoch timestamp at which a person transferred an amount
6. redeemed_at : epoch timestamp at which a person redeemed an item

```python
df = pd.read_json('SampleData(15Nov2018_to_13Sep2019 (2).json',orient='index')
df
```

```python
df.dtypes
```

# 1. Converting the epoch timestamp of each redemption to a proper Date format (dd-mm-yyyy)


```python
from datetime import datetime
df['redeemed_date']= pd.to_datetime(df.redeemed_at).dt.strftime('%d-%m-%Y')
df
```

```python
# Extracting Month and Year in another column to create the overview.
df['redeemed_month_year']=df['redeemed_at'].dt.strftime('%b %Y')
df
```

```python
df.sort_values(by='redeemed_month_year')
df
```

```python
print (df.isnull().sum())
```

# 2. Creating an overview of No. of transactions per User for the period between Nov 2018 to September 2019.

```python
df2 = pd.pivot_table(df, values='paid_amount', index=['user_id'],columns=['redeemed_month_year'], aggfunc='count',fill_value=0)
print(df2.to_string())
```

```python
df2.sort_index(axis='rows')
df2
```
